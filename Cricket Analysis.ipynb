```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 5, figsize=(20, 8))  # 2 rows, 5 columns

for i, ax in enumerate(axs.flat):
    ax.plot(Salary[i], c='tab:blue', ls='--', marker='o', ms=6)
    ax.set_title(Players[i])
    ax.set_xticks(range(len(Seasons)))
    ax.set_xticklabels(Seasons, rotation='vertical')
    ax.set_ylim(0, max(max(s) for s in Salary) * 1.1)
    ax.grid(True, linestyle=':', linewidth=0.5)

fig.suptitle("Player Salaries Over Seasons", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Find top 3 players by average salary
avg_salaries = [sum(s)/len(s) for s in Salary]
top_indices = sorted(range(len(avg_salaries)), key=lambda i: avg_salaries[i], reverse=True)[:3]

for i in top_indices:
    plt.plot(Salary[i], label=Players[i], marker='o', ms=6)

plt.xticks(range(len(Seasons)), Seasons, rotation='vertical')
plt.legend()
plt.title("Top 3 Highest Paid Players - Salary Over Seasons")
plt.xlabel("Season")
plt.ylabel("Salary (in millions)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
